```python
a["TotalCharges"]=a["TotalCharges"].replace(" ","0")
a["TotalCharges"]=a["TotalCharges"].astype("float")
a.info()
```

```python
def chang(var):
    if var==1:
        return "yes"
    else:
        return "no"
a["SeniorCitizen"]=a["SeniorCitizen"].apply(chang)
a.info()
```

#Now our data is clear now we look the amount/percentage of churn people

```python
plt.figure(figsize=(4,4))
b=sns.countplot(x="Churn",data=a)
plt.title("Count Of Churn Out",fontsize=12)
b.bar_label(b.containers[0])
plt.show()
```

```python
plt.figure(figsize=(3,3))
gb=a.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb["Churn"],labels=gb.index,autopct="%1.2f%%")
plt.title("% Of Churn Out")
#a.bar_label(a.container[0])
plt.show()
print(gb)
```

```python
# List of columns to plot
columns_to_plot = ['SeniorCitizen',"Partner", "Dependents","PhoneService"
                   ,"MultipleLines","InternetService","OnlineSecurity","OnlineBackup"
                   ,"DeviceProtection","TechSupport","StreamingTV","StreamingMovies","Contract"
                   ,"PaperlessBilling","PaymentMethod"]

# Create subplots
fig, axes = plt.subplots(nrows=(len(columns_to_plot) + 2) // 3, ncols=3, figsize=(12, (len(columns_to_plot) + 2) // 3 * 4))

for ax, column in zip(axes.flatten(), columns_to_plot):
    sns.countplot(x=column, data=a, hue='Churn', ax=ax)
    ax.set_title(f'Count of Churn on Basis of {column}')

# Remove any empty subplots
for i in range(len(columns_to_plot), len(axes.flatten())):
    fig.delaxes(axes.flatten()[i])

plt.tight_layout()
plt.show()
```
